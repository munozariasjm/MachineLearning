# Feature Representation: Transforming Data

Estamos buscando una función $\Phi$ que facilite el trabajo (como hacer los datos separables a travéz del origen), esta función es $$\Phi: \mathbb{R}^d \longmapsto \mathbb{R}^D$$
Generalmente $D\geq d$ pero no debe ser así. La idea detrás es que si se tiene un separador que no va a través del origen en $d$ dimensiones, se puede construir un separador que si pase por el origen en $d+1$ dimensiones.
